# Cartopy Projections

## Selection of the correct map projection is very important for visualization of data within the geosciences. Different projections fundamentally work in different ways and therefore work better  for different purposes and areas of the globe

### Type of Projections

When evaluating different projections, here is some of the most frequent terminology that should be taken into account

##### Projection Shape

- Cylindrical -> typically better in tropical regions
- Conic -> typically better in the mid-latitudes
- Azimuthal -> typically better in polar regions

##### Projection Metric Properties

- Conformal -> perserves projection at a local scale
- Equal Area -> perserves projection area
- Equadistant -> perserves projection distance

Remember, the ultimate goal of choosing a projection is to maintain the shape of what ever is being displayed.

### Example Projections

##### Mercador -> Cylindrical & Conformal (recommendend in the tropics at a local scale)

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cf

plt.figure(figsize=(6, 3))
ax = plt.axes(projection=ccrs.Mercator())
ax.add_feature(cf.COASTLINE)
ax.gridlines()
```

##### Lambert Conformal -> Conic & Conformal (recommendend in the midlatitudes at a local scale)

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cf

plt.figure(figsize=(6, 3))
ax = plt.axes(projection=ccrs.LambertConformal())
ax.add_feature(cf.COASTLINE)
ax.gridlines()
```

##### Sinusoidal -> Cylindrical & Equal Area (recommendend in the tropics when trying to perserve area)

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cf

plt.figure(figsize=(6, 3))
ax = plt.axes(projection=ccrs.Sinusoidal())
ax.add_feature(cf.COASTLINE)
ax.gridlines()
```
