# Cartopy

## Cartopy is a useful tool in displaying geospatial data and providing a base for mapping this information, essential for the geosciences. Built on top of Matplotlib, it makes data visualization straightforward and easy.

##### Before working with cartopy, instaliation of the package is necessary. Here are three of the main subpackages that are typically used:

- cartopy coordinate reference system -> allows cartopy to utilize coordinate reference systems (i.e. Lambert, Mercator, etc.)

- cartopy features -> provides geographical features for cartopy plotting (i.e. rivers, geopolitical boundaries, coastlines, etc.)

- cartopy shapereader -> allows for cartopy to utilize shapefiles

```python
import cartopy                                      # base cartopy package
import cartopy.crs as ccrs                          # cartopy coordinate reference system
import cartopy.feature as cf                        # cartopy features
import cartopy.io.shapereader as shpreader          # cartopy shapereader
```

Furthermore, since cartopy uses a matplotlib base, importing subpackages from that package is important as well

```python
import matplotlib.pyplot as plt 
```

### Cartopy Basics

Usually the first step in utilizing cartopy is to use a coordinate reference system

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator())
```

As you can see, nothing is plotted on this figure. This is because there no features or data being displayed.

Let's add a coastlines feature to this map

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator())

ax.add_feature(cf.COASTLINE)
```

We can also shift the center longitude and latitude (for some coordinate reference systems) of the map if wanted

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator(central_longitude = 270))
ax.add_feature(cf.COASTLINE)
```

Furthermore, we can set the latitudinal and longitudinal bounds of the map using set_extent

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator())
ax.add_feature(cf.COASTLINE)

ax.set_extent([-130, -50, 20, 45])
```

Now the map is a little harder to read. Let's add more features to this map

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator())
ax.add_feature(cf.COASTLINE)
ax.set_extent([-130, -50, 22, 45])

ax.add_feature(cf.LAND)
ax.add_feature(cf.OCEAN)
ax.add_feature(cf.LAKES)
ax.add_feature(cf.BORDERS)
ax.add_feature(cf.STATES)
```

Now we have a map of the United States that is easy to read and ready (with some slight modifications as warented) to plot our desired data

More information on cartopy basics can be found here: https://scitools.org.uk/cartopy/docs/latest/getting_started/index.html

### Shapefiles with Cartopy

Cartopy also can be used to read in shapefiles stored either locally or on the internet

The following is an example of a cartopy shapereader reading in an online shapefile of United States counties

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.Mercator())

ax.set_extent([-130, -50, 22, 45])
ax.add_geometries(shpreader.Reader('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/US-counties.geojson').geometries(), 
                  ccrs.PlateCarree())

ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.LAKES)
ax.add_feature(cf.BORDERS)
ax.add_feature(cf.STATES)
```

More information about cartopy and shapefiles can be found here: https://scitools.org.uk/cartopy/docs/latest/tutorials/using_the_shapereader.html
